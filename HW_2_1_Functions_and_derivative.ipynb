1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.



```python
import numpy as np
import matplotlib.pyplot as plt
import sympy as sp
```

```python
x = np.linspace(0, 10, 20)
f = 10 * x - 5

df = sp.diff(10 * sp.symbols('x') - 5)

plt.figure(figsize=(20, 10))
plt.plot(x, f, label='f(x) = 10*x-5', color='blue')
plt.plot(x, np.full_like(x, df), label="f'(x) = 10", color='green')

x_point = 3
f_point = 10 * x_point - 5
df_point = 10
tangent_line = df_point * (x - x_point) + f_point

plt.scatter(x_point, f_point, color='red')
plt.plot(x, tangent_line, '--', color='red', label='Tangent at x = 3')

plt.title("Графік функції та її похідної")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.grid(True)

plt.show()
```

**Висновки:**
1. Функція є лінійною і відображена на графіку синьою лінією
2. Похідна функції становить 10 що є константою, тобто з кожною додатковою годиною навчання оцінка буде підвищуватись на 10
3. Оскільки функція лінійна, її дотична в будь-якій точці збігається з самою функцією, що зображено червоною пунктирною лінією, і швидкість росту оцінки не збільшується з часом

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.



```python
def plot_function_and_derivative(f, df, x_range=(0, 10), x_point=3, num_points=100):
    x = np.linspace(x_range[0], x_range[1], num_points)
    y = f(x)
    dy = df(x)

    # Точка дотику
    f_point = f(x_point)
    df_point = df(x_point)
    tangent = df_point * (x - x_point) + f_point

    # Побудова графіків
    plt.figure(figsize=(12, 6))
    plt.plot(x, y, label='f(x)', color='blue')
    plt.plot(x, dy, label="f'(x)", color='green')
    plt.plot(x, tangent, '--', label=f'Tangent at x = {x_point}', color='red')
    plt.scatter(x_point, f_point, color='red')  # Точка дотику

    plt.title("Графік функції, її похідної та дотичної")
    plt.xlabel("x")
    plt.ylabel("y")
    plt.legend()
    plt.grid(True)
    plt.show()
```

```python
x_sym = sp.symbols('x')
f_expr = sp.sin(x_sym)
df_expr = sp.diff(f_expr, x_sym)

f_num = sp.lambdify(x_sym, f_expr, 'numpy')
df_num = sp.lambdify(x_sym, df_expr, 'numpy')

plot_function_and_derivative(
    f_num,
    df_num,
    x_range=(0, 10),
    x_point=2
)
```

```python
x_sym = sp.symbols('x')
f_expr = sp.ln(x_sym)
df_expr = sp.diff(f_expr, x_sym)

f_num = sp.lambdify(x_sym, f_expr, 'numpy')
df_num = sp.lambdify(x_sym, df_expr, 'numpy')

plot_function_and_derivative(
    f_num,
    df_num,
    x_range=(0, 10),
    x_point=2
)
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x_sym = sp.symbols('x')
f_expr = x_sym**2 - 5*x_sym + 1
df_expr = sp.diff(f_expr, x_sym)

f_num = sp.lambdify(x_sym, f_expr, 'numpy')
df_num = sp.lambdify(x_sym, df_expr, 'numpy')

plot_function_and_derivative(
    f_num,
    df_num,
    x_range=(0, 100),
    x_point=25
)
```

Спостереження:

- Синя - графік квадратичної функції, показує спочатку повільне зростання витрат, а потім різке збільшення, що означає при збільшенні кількості одиниць продукції вартість виробництва різко зростає

- Зелена - графік похідної функції що відображає ефективність виробництва, яке є дуже низьке

- Червона пунктирна - дотична в точці 25, та відображає напрямок і швидкість змін витрат в цей момент, вона має досить крутий нахил і вказує на те що в цій точці витрати досить швидко ростуть і якщо обрати більшу точку дотичної то витрати будуть ще більше збільшуватись
